# GDP AND LIFE EXPECTANCY OF SIX COUNTRIES

### Load and inspect Data

```python
#load and inspect dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

all_data = pd.read_csv('all_data.csv')
print(all_data.head())
print(all_data.columns)
print(all_data.dtypes)
print(all_data.value_counts().sum())

#check for duplicates
duplicates = all_data.duplicated()
print(duplicates.value_counts())

#unique countries in datatset
print('The countries included in this dataset are:')
Country = all_data.groupby('Country')
for key, value in Country:
    print(key)
    
#unique years in dataset
print('The unique years in this dataset are:')
Year = all_data.groupby('Year')
for key, value in Year:
    print(key)
```

### Data Wrangling

```python
##Data Wrangling
#rename column
all_data.rename(columns = {'Life expectancy at birth (years)': 'LifeExpectancy'}, inplace=True)
print(all_data.columns)
```

### Data Analysis

```python
##Data Analysis
#summary statistics on dataset
print(all_data.describe())
print(all_data.info())

#summary statistics of Life Expectancy for each country
#What is the average life expectancy in these nations?
all_data.groupby("Country")["LifeExpectancy"].describe()
```

#### Germany had the highest Life Expectancy mean of approximately 79.7 years with Zimbabwe with the lowest Life Expectancy mean of 50 years.

```python
#What is the average GDP (USD) in these nations?
all_data.groupby("Country")["GDP"].describe()
```

#### United States of America had the highest GDP mean while Zimbabwe had the lowest GDP mean.

### Data Visualization

```python
##Data Visualization
#Has life expectancy increased over time in the six nations?
for country in all_data["Country"].unique():
    plt.plot(all_data[all_data["Country"] == country]["Year"], all_data[all_data["Country"] == country]["LifeExpectancy"], label = country, marker = "o")

plt.xlabel("Year")
plt.ylabel("Life Expectancy at birth (years)")
plt.title("Life Expectancy of six countries over time")
plt.legend()
plt.show()
plt.clf()
```

```python
#Looking closely at Life expectancy over time in each country
LifeExpectancy_facegrid = sns.FacetGrid(all_data, col= "Country", col_wrap=3,
                               hue = "Country", sharey = False)

LifeExpectancy_facegrid = (LifeExpectancy_facegrid.map(sns.lineplot, "Year", "LifeExpectancy")
                  .add_legend()
                  .set_axis_labels("Year","Life expectancy at birth (years)"))

LifeExpectancy_facegrid;
```

#### All six countries have experienced increase in life expectancy over the years, 2000 through 2015. Although Germany had the highest life expectancy in 2015, Zimbabwe had the highest increase of 15 years (45 - 60 years), while Mexico had the shortest increase of 1.5years (75 - 76.5years).

```python
#Has GDP increased over time in the six nations?
for country in all_data["Country"].unique():
    plt.plot(all_data[all_data["Country"] == country]["Year"], all_data[all_data["Country"] == country]["GDP"], label = country, marker = "o")

plt.xlabel("Year")
plt.ylabel("GDP (USD)")
plt.title("GDP of six countries over time")
plt.legend(loc=1)
plt.show()
plt.clf()
```

```python
GDP_facegrid = sns.FacetGrid(all_data, col= "Country", col_wrap=3,
                               hue = "Country", sharey = False)

GDP_facegrid = (GDP_facegrid.map(sns.lineplot, "Year", "GDP")
                  .add_legend()
                  .set_axis_labels("Year","GDP (USD)"))

GDP_facegrid;
```

#### All six of the countries experienced an increase in GDP over the years. Chile and Germany had the hight growth in GDP. China and USA had the most sustained growth in GDP.

```python
#Is there a correlation between GDP & Life Expectancy?
sns.scatterplot(x = "LifeExpectancy", y = "GDP", hue = "Country", data = all_data)
plt.show()
plt.clf()
```

```python
LifeExpectancy_GDP_facegrid = sns.FacetGrid(all_data, col="Country", col_wrap=3, hue ="Country",  sharey = False, sharex = False)
LifeExpectancy_GDP = (LifeExpectancy_GDP_facegrid.map(sns.scatterplot, "LifeExpectancy", "GDP").add_legend()
             .set_axis_labels("Life expectancy at birth (years)", "GDP (USD)"));
LifeExpectancy_GDP_facegrid;
```

#### We can say there is a positive correlation between the GDP and Life Expectancy of each of the six countries.

```python
#What is the distribution of that Life Expectancy in all the countries?
LifeExpectancy_Dist_facegrid = sns.FacetGrid(all_data, col="Country", col_wrap=3,
                      hue = "Country", sharey = False, sharex = False)
LifeExpectancy_Dist_facegrid = (LifeExpectancy_Dist_facegrid.map(sns.histplot, "LifeExpectancy", kde=True, bins=10)
         .add_legend())

LifeExpectancy_Dist_facegrid.fig.subplots_adjust(top=0.9)
LifeExpectancy_Dist_facegrid.fig.suptitle('Distribution of Life expectancy in each of the six countries')
plt.show()
plt.clf()
```
